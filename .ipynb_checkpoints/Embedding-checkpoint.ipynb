## ResNet Embedding

Uses [this tutorial](https://www.activeloop.ai/resources/1E3oKP83o4KXIFChCesv7r/generate-image-embeddings-using-a-pre-trained-cnn-and-store-them-in-hub/)

```python
# fetch pretrained model

model = torch.hub.load('pytorch/vision:v0.10.0', 'resnet18', pretrained=True)
```

```python
# Select the desired layer, one before it classifies
layer = model._modules.get('avgpool')

#use register_forward_hook module to get embeddings
def copy_embeddings(m, i, o):

    """Copy embeddings from the penultimate layer.

    """

    o = o[:, :, 0, 0].detach().numpy().tolist()

    outputs.append(o)

outputs = []

# attach hook to the penulimate layer
_ = layer.register_forward_hook(copy_embeddings)
```

```python
model.eval() # Inference mode

# Generate image embeddings for all images in dloader and save them in the list outputs
for X, y in dloader:
    _ = model(X)

print(len(outputs)) 
```

```python
# flatten list of embeddings to remove batches
list_embeddings = [item for sublist in outputs for item in sublist]

print(len(list_embeddings)) # should return same length as dloader
print(np.array(list_embeddings[0]).shape) #returns (512,)
```

# Clustering - nearest neighbor

```python
from tqdm import tqdm

num_clusters = 12  # Number of clusters.
k_neighbours = 550 # Number of neighbours to consider during cluster learning.
tune_encoder_during_clustering = False  # Freeze the encoder in the cluster learning.
```

```python
random.seed(30)

neighbours = []
batch_size = 500
num_batches = len(list_embeddings) // batch_size
for batch_idx in tqdm(range(num_batches)):
    start_idx = batch_idx * batch_size
    end_idx = start_idx + batch_size
    current_batch = list_embeddings[start_idx:end_idx]
    # Compute the dot similarity.
    similarities = tf.linalg.matmul(current_batch, list_embeddings, transpose_b=True)
    #print(similarities)
    # Get the indices of most similar vectors.
    _, indices = tf.math.top_k(similarities, k=k_neighbours + 1, sorted=True)
    
    
    # Add the indices to the neighbours.
    neighbours.append(indices[..., 1:])

neighbours = np.reshape(np.array(neighbours), (-1, k_neighbours))
```

## Evaluation

```python
def evaluate_images_sim(image_datasets, neighbours, example_num = 5, neighborhood_max = False):
    """ Returns evaluation metric for image similarity based on number predicted in neighborhood
    that are the same class as input image versus the total number predicted in the neighborhood
    
    INPUT
    ------
    
    OUTPUT
    ------
    
    """
    
    from statistics import mean
    
    target_class = 0
    base_photos = {}
    class_list = []
    class_indicies = []
    
    #iterate through images and find some matching each class
    for i in range(len(image_datasets)):
        class_label = image_datasets[i][1]
    
        #if the photo has target class
        if class_label == target_class:
            base_idx = []
            
            #for amount of examples chosen append that many photo indicies
            for _ in range(example_num):
                #base_photos[f"{target_class}"] = image_datasets[i][0]
                base_idx.append(i)
                i += 1
            
            #append current classes' indicies and label to lists
            class_list.append(target_class)
            class_indicies.append(base_idx)
            #increase target class and loop statement again
            target_class += 1
            continue     
            
    #empty dict for accuracy means        
    class_mean = {}
    
    #iterate number of classes
    for i in range(len(class_indicies)):
    
        class_accuracy = []
        
        #for every photo idx within each class
        for j in range(len(class_indicies[i])):
        
             #only take unique neighbours
            unique_neighbours = []
            for neighbour in neighbours[j]:
                if neighbour != class_indicies[i][j]:
                     unique_neighbours.append(neighbour)
             
            #get the neighbour classes
            neighbour_classes = []
            for neighbour in unique_neighbours:
                neighbour_classes.append(image_datasets[neighbour][1])
        
            #adjust how many values are considered neighbours
            if bool(neighborhood_max):
                neighbour_classes = neighbour_classes[:neighborhood_max]
        
            #append accuracy for each photo
            class_accuracy.append(neighbour_classes.count(class_list[i]) / len(neighbour_classes))
        
        #get mean for all accuracies and save to dict
        class_mean[class_list[i]] = round(mean(class_accuracy), 4)
        
    return class_indicies, class_mean
```

```python
photos_used, class_mean = evaluate_images_sim(image_datasets, neighbours, example_num = 10)

class_mean
```

```python
neighbour_limits = range(1, 300, 20)
class_mean_plot = {}

for i in range(16):
    class_mean_plot[i] = []

for limit in neighbour_limits:
    photos_used, class_mean = evaluate_images_sim(image_datasets, neighbours, example_num = 5, neighborhood_max = limit)
    
    for key, value in class_mean.items():
        class_mean_plot[key].append(value)
            
```

```python
import matplotlib.ticker as mticker

fig, ax = plt.subplots()

for i in class_mean_plot.keys():
    plt.plot(class_mean_plot[i])

    ax.xaxis.set_major_locator(mticker.FixedLocator(range(len(class_mean_plot[0]))))
    ax.set_xticklabels(neighbour_limits)
    ax.set_xlabel('Number of Neighbours')
        
    ax.set_ylabel('Accuracy %')
    
plt.savefig('evaluation/version_4/neighbor_threshold.png')
```
